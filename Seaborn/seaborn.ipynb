## Seaborn Styling
- When styling
our figures, we should take into consideration our audience and the context.
- As questions like: 
    - Is your visualization part of a report & needs to convey specific information?
    - Or is it part of a presentation?
    - Or is your visualization meant as its own stand-alone, with no narrator in front of it.
- People are more likely to concentrate on beautiful and attractive visualizations rather than dull plots thus styling can be considered as a vital component of data visualization.

#### 1) Built-in Themes

- Seaborn supports various themes that can make styling the plots really easy and save a lot of time.
- There are 5 built-in themes to style its plots:
    - Darkgrid.
    - Whitegrid.
    - dark.
    - white.
    - ticks.
- Seaborn defaults to using the white theme for its plots.
- <b>sns.set_style()</b> method is used to set the themes that we want to use in different plots.
<u>e.g sns.set_style('dark')</u>

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tips=sns.load_dataset('tips')

sns.set_style('darkgrid')

sns.stripplot(x='day', y='total_bill', data=tips)
plt.savefig('FirstPlot.png')
plt.show()
```

#### 3) Despine
- Spines are the the borders of the figure that contains the visualization.
- By default, the visualizations have 4 spines.
- <b>despine()</b> method removes the top and the right spines.
- To remove all the spines, we can pass argument to the despine method, <b>despine(bottom=True, left=True)</b>.
- despine() function is called after you have called the plot.

#### Scaling
- Seaborn has four presets which sets the size of the plot & allow you to customize your figure depending on how you want to present it.
- Inorder of relative size they are:
    - paper
    - notebook
    - talk
    - poster
- The <b>notebook</b> style is the default.
- <b>set_context()</b> method is used for scaling plots.
- <b>paper</b> is the smallest context.
- <b>poster</b> is the largest context.

```python
tips=sns.load_dataset('tips')

sns.set_style('ticks')

sns.stripplot(x='day', y='total_bill', data=tips)
sns.despine(bottom=True, left=True)
sns.set_context('talk')
plt.savefig('FirstPlot.png')
plt.show()
```

## Seaborn Plotting functions

### Figure level functions and Axes-level functions

#### Axes Level Functions.
- They plot data onto a single matplotlib.pyplot.Axes object, which is the return value of the function.

#### Figure Level Functions.
- Figure-level function is an interface with matplotlib through a seaborn object, usually a FacetGrid, that manages the figure. 
- Each module has a single figure-level function, which offers a unitary interface to its various axes-level functions.
- For example, **relplot()** is the figure-level function for the relational module. 
- Its default behavior is to draw a scatter plot, using the same code as **scatterplot()** behind the scenes.

### Distribution plots

Suitable for comparing range and distribution for groups of numerical data

- displot -Figure-level interface for drawing distribution plots onto a FacetGrid.
- histplot -Plot univariate or bivariate histograms to show distributions of datasets.
- kdeplot -Plot univariate or bivariate distributions using kernel density estimation.
- ecdfplot - Plot empirical cumulative distribution functions.ECDF represents the proportion or count of observations falling below each unique value in a dataset.
- rugplot -Plot marginal distributions by drawing ticks along the x and y axes

##### Displot
- Uses sns.displot() method.
- provides access to several approaches for visualizing the univariate or bivariate distribution of data, including subsets of data defined by semantic mapping and faceting across multiple subplots. 
- The kind parameter selects the approach to use:

    - histplot() (with kind="hist"; the default)

    - kdeplot() (with kind="kde")

    - ecdfplot() (with kind="ecdf"; univariate-only)
 
- The distplot() function combines the matplotlib hist function with the seaborn kdeplot() and rugplot() functions. 
- Docs: https://seaborn.pydata.org/generated/seaborn.displot.html

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
# The default plot kind is a histogram:
sns.displot(data=titanic, x='age', kind='kde')
plt.title("Age Distribution")
plt.show()
```

```python
# The default plot kind is a histogram:
sns.displot(data=titanic, x='age', kde=True, rug=True)
plt.title("Age Distribution")
plt.show()
```

##### ecdf plot
- Short for <b>empirical distribution function</b>
- In statistics, an empirical distribution function is the distribution function associated with the empirical measure of a sample. 
- An ECDF represents the proportion or count of observations falling below each unique value in a dataset. 
- Compared to a histogram or density plot, it has the advantage that each observation is visualized directly, meaning that there are no binning or smoothing parameters that need to be adjusted.
- Docs: https://seaborn.pydata.org/generated/seaborn.ecdfplot.html

###### Example
Say if you have plotted an ECDF plot for the entire class students math exam score and you wanted to know what is the score required by a student to reach the top 10% math scorer in the class. You could just look at the value corresponding to the 90% in ECDF plot and straight away get the score value.

```python
penguins=sns.load_dataset('penguins')
penguins.head()
```

```python
sns.ecdfplot(data=penguins, x="flipper_length_mm")
plt.axhline(0.85)
plt.axvline(218)
plt.show()
```

```python
# This can be arrived at using numpy's percentile function.
flipper_length = np.array(penguins['flipper_length_mm'].dropna())
np.percentile(flipper_length,[50, 85], axis=0)
```

The ECDF plot could be used to answer different questions. For example, one question could be what is the flipper length value in which 50% of the sample is below that value and 50% above that value.
<br><br>
Similarly, you could identify the value corresponding to 85% which shows 85% of the observation is below that value and 15% observations are above that value.
- Read more abou ECDF: https://towardsdatascience.com/how-to-generate-ecdf-plot-using-python-and-r-247ef81fbf3f

##### KDE Plots
- KDE stands for Kernel Density Estimator.
- Use the sns.kdeplot() function.
- A KDE plot gives us the sense of an univariate dataset as a curve give us the shape and the distribution of the dataset.
- KDE plots are prefarable to histograms because depending on how you group the data into bins & the width of the bins, you can draw wildly different conclusions about the shape of the data.
- Using KDE plot can mitigate these issues because they smooth the datasets, allowing us to generalize the shape of the data.
- Docs: https://seaborn.pydata.org/generated/seaborn.kdeplot.html

```python
sns.kdeplot(data=penguins, x='flipper_length_mm', shade=True)
plt.show()
```

In the KDE plot above we can be able to easily read various statistical aspects of the dataset.
- modality -> Number of distinct peaks. (2). Thus, Pi-modal
- Skewness -> Right-sided skew (Positive Skew)

```python
x=penguins['flipper_length_mm'].dropna()
mean=np.mean(x)
median=np.median(x)

# N/B Refused to give median because there were null values 
```

```python
# In a Positive skewed distribution the mean is larger than the median
sns.kdeplot(data=penguins, x="flipper_length_mm")
plt.axvline(median, color='red')
plt.axvline(mean, color='purple')
plt.show()
```

### Categorical plots
* catplot  Figure-level interface for drawing categorical plots onto a FacetGrid.
* stripplot  Draw a scatterplot where one variable is categorical.
* swarmplot -Draw a categorical scatterplot with non-overlapping points.
* boxplot  -Draw a box plot to show distributions with respect to categories.
* violinplot - Draw a combination of boxplot and kernel density estimate.
* boxenplot  -Draw an enhanced box plot for larger datasets.
* pointplot  -Show point estimates and confidence intervals using scatter plot glyphs.
* barplot   -Show point estimates and confidence intervals as rectangular bars.
* countplot  -Show the counts of observations in each categorical bin using bars.

##### Catplot (figure level function)
- This function provides access to several axes-level functions that show the relationship between a numerical and one or more categorical variables using one of several visual representations.
- The kind parameter selects the underlying axes-level function to use:

- Categorical scatterplots:

    - stripplot() (with kind="strip"; the default)

    - swarmplot() (with kind="swarm")

- Categorical distribution plots:

    - boxplot() (with kind="box")

    - violinplot() (with kind="violin")

    - boxenplot() (with kind="boxen")

- Categorical estimate plots:

    - pointplot() (with kind="point")

    - barplot() (with kind="bar")

    - countplot() (with kind="count")

- Docs: https://seaborn.pydata.org/generated/seaborn.catplot.html

```python
exercise = sns.load_dataset('exercise')
sns.catplot(x="time", y="pulse", hue="kind", data=exercise, kind='bar', ci=None)
plt.show()
# By default, Seaborn will place error bars on each bar when you use the barplot.
# Error bars visually indicate the range of the values that might be expected for that bar.
```

##### Boxplot
- A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. 
- The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be outliers using a method that is a function of the inter-quartile range.
- The longer the boxplot the bigger the spread of the distribution.
- There are different ways to plot the wwhiskers of a boxplot:
    - Extend whiskers to the min and max values.
    - Extend them one standard deviation from the mean.
    - The most common used way is to extend them 1.5 times the interquartile range from  the 1st and 3rd quartiles.
- Docs: https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
sns.boxplot(data=tips)
plt.show()
```

```python
sns.boxplot(data=tips['total_bill'])
plt.show()
```

- Upper side of the box is the q3
- Middle line in the box is the median
- lower side of the box is the q1
- values that have faal beyond th whisker are the outliers

##### Barplot
- Used sns.barplot() function.
- A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars.
- Docs: https://seaborn.pydata.org/generated/seaborn.barplot.html

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=tips,ci=None)
plt.show()
```

##### Violin Plots
- A violin plot plays a similar role as a box and whisker plot. 
- It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. 
- Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.
Docs: https://seaborn.pydata.org/generated/seaborn.violinplot.html

```python
sns.violinplot(x=tips["total_bill"])
plt.show()
```

- Yellowish dot is the median
- Thick line in the middle is the inteerquartile range
- Thin line is the whisker 
- Has a right skew distribution at the top as it has a longer tail to the right.

### Regression plots
- Creates a regression line between 2 parameters and helps to visualize their linear relationships.

- lmplot -Plot data and regression model fits across a FacetGrid.
- regplot -Plot data and a linear regression model fit.
- residplot -Plot the residuals of a linear regression.

##### lmplot (figure level function)
- This function combines regplot() and FacetGrid.
- It is intended as a convenient interface to fit regression models across conditional subsets of a dataset.
- The regplot() and lmplot() functions are closely related, but the former is an axes-level function while the latter is a figure-level function that combines regplot() and FacetGrid
- Docs: https://seaborn.pydata.org/generated/seaborn.lmplot.html

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
plt.show()
```

Shaded region represents the confidence interval, given a similar dataset, the line of best fit(regression line) will fall withing the region

```python
sns.lmplot(x="total_bill", y="tip", data=tips, ci=None)
plt.show()
```

##### Regplot

### Matrix plots

A matrix plot is an array of scatterplots. effective when you are only interested in relationships among certain pairings of variables.

- heatmap -Plot rectangular data as a color-encoded matrix.
- clustermap -Plot a matrix dataset as a hierarchically-clustered heatmap.



##### Heatmap
- Plots rectangular data as a color-encoded matrix.
- Better visualize the volume of locations/events within a dataset and assist in directing viewers towards areas on data visualizations that matter most.
Docs: https://seaborn.pydata.org/generated/seaborn.heatmap.html

```python
random=np.random.rand(10,12)
sns.heatmap(random)
plt.show()
```
