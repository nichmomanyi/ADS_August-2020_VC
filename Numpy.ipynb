#### Numpy Arrays

* NumPy are building blocks of most of the Numpy operations
* It’s a table of elements (numbers) of the same type, indexed by a tuple of positive intergers
* Numpy dimensions are called **axes** and number of axes is **rank**
* Numpy’s array class is called **ndarray**
* They store the same type of data

```python
x=[1,2,3,4,5]

# creating a numpy array from a python list
nums=np.array(x)
nums
```

```python
nums.dtype
```

```python
# Getting the dimension
nums.ndim
```
